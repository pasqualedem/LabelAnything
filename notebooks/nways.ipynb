```python
import pandas as pd

data = {
    'N-way': [1, 2, 3, 4, 5, 10, 15, 20],
    'DCAMA': [50.9, 31.7, 24.2, 21.1, 16.7, 10.8, 6.5, 4.7],
    'LA': [39.7, 31.9, 28.6, 26.5, 25.1, 20.7, 15.0, 10.8]
}

df = pd.DataFrame(data)
print(df)
```

```python
ratio = 2.5
size = 4

ax = df.plot(
    x='N-way',
    y=['DCAMA', 'LA'],
    kind='line',
    marker='o',
    title=None,
    color=['blue', 'green'],
    figsize=(size * ratio, size)
)
ax.grid(True)
# Set custom linestyles for each line
ax.lines[0].set_linestyle('dashed')  # DCAMA
ax.lines[1].set_linestyle('solid')   # LA
ax.set_xticks(df['N-way'])

displacements_dcama = [
    (-0.5, -3.3),  # N-way = 1
    (-0.5, -3.3),  # N-way = 2
    (-0.5, -3.3),  # N-way = 3
    (-0.5, -3.3),  # N-way = 4
    (-0.5, -3.3),  # N-way = 5
    (-0.5, -3.3), # N-way = 10
    (-0.5, -3.3), # N-way = 15
    (-0.5, -2.3)  # N-way = 20
]
displacements_la = [
    (-0.1, +1),  # N-way = 1
    (+0.5, +1),  # N-way = 2
    (+0.5, +1),  # N-way = 3
    (+0.5, +1),  # N-way = 4
    (+0.6, +1),  # N-way = 5
    (+0.5, +1), # N-way = 10
    (+0.5, +1), # N-way = 15
    (+0.5, +1)  # N-way = 20
]
ax.set_xlim(left=-0.25, right=21.2)

for i, row in df.iterrows():
    ax.text(row['N-way'] + displacements_dcama[i][0], row['DCAMA'] + displacements_dcama[i][1], f"{row['DCAMA']}", va='bottom', ha='center')
    ax.text(row['N-way'] + displacements_la[i][0], row['LA'] + displacements_la[i][1], f"{row['LA']}", va='bottom', ha='center')
    
    ax.figure.savefig("nways.pdf", format="pdf", bbox_inches="tight")
```
